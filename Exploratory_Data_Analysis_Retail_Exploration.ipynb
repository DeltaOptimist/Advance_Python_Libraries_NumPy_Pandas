### Correlation for Profit vs Discount

```python
fig,ax=plt.subplots(figsize=(20,8))
ax.scatter(df_data['Sales'],df_data['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df_data)
plt.legend()
plt.show()
```

### Resultant Observation from above <br>

### Looking Corelation in Profit vs Quantity

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df_data)
plt.legend()
plt.show()
```

```python
df_data.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['pink','blue'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

### Resultant Observation from above  <br>
- Profit and sales are maximum in consumer segment and minimum in Home Office segment

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region',y='Sales',data=df_data,hue='Segment',order=df_data['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=15)
plt.show()
```

###Resultant Observation from above  <br>
- Segment wise sales are almost same in every region

```python
df_data.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['blue','red'],figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

### Resultant Observation from above <br>
- Profit and sales are maximum in west region and minimum in south region

```python
ps = df_data.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['blue','orange'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('States')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

### Resultant Observation from above  <br>
- high profit is for california, new york<br>
- loss is for texas, pennsylvania, Ohio

```python
t_states = df_data['State'].value_counts().nlargest(10)
t_states
```

```python
df_data.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['yellow','purple'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

### Resultant Observation from above  <br>
#### As a business manager, try to find out the weak areas where you can work to make more profit?<br>

- Technology and Office Supplies have high profit.<br>
- Furniture have less profit<br>

```python
ps = df_data.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','lightblue'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

### Resultant Observation from above <br>

- Phones sub-category have high sales.<br>
- chairs have high sales but less profit compared to phones<br>
- Tables and Bookmarks sub-categories facing huge loss<br>
